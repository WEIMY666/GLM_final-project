```r
train_data = read.table("D:/GLM final/2001_SPECTF Heart Data Set/2001_SPECTF Heart Data Set/SPECTF.train.txt",sep=",")
valid_data = read.table("D:/GLM final/2001_SPECTF Heart Data Set/2001_SPECTF Heart Data Set/SPECTF.test.txt",sep=",")
```

```r
train_data = data.matrix(train_data)
valid_data = data.matrix(valid_data)
```

```r
x_train = train_data[,2:45]
y_train = train_data[,1]
x_test = valid_data[,2:45]
y_test = valid_data[,1]
```

```r
#install.packages("glmnet")
library(glmnet)
```

```r
logistic_lasso  = glmnet(x_train,y_train,family = "binomial",alpha = 1)
```

```r
plot(logistic_lasso)
cv = cv.glmnet(x_train,y_train,nfolds = 10,standardize = TRUE,family = "binomial",alpha = 1)
plot(cv)
bestlambda = cv$lambda.min

fit = glmnet(x_train,y_train,family = "binomial",lambda = bestlambda,alpha = 1)
```

```r
pred = predict(fit,newx = x_test,type = "class")
```

```r
T=table(y_test,pred,dnn=c("真实值","预测值"))    #输出混淆矩阵
T
accuracy = (T[1,1]+T[2,2])/187
accuracy
```
